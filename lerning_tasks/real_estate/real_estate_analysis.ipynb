## Предобработка данных

### Работа с дубликатами

```python
df.duplicated().sum()
```

Дубликатов нет! Странно, но радует.

### Работа со строковыми значениями и датами

```python
# переименовываем столбцы 
df = df.rename(columns={
    'airports_nearest' : 'airports_distance_meters',
    'studio' : 'is_studio',
    'cityCenters_nearest' : 'city_center_distance_meters',
    'parks_nearest' : 'parks_distance_meters',
    'ponds_nearest' : 'ponds_distance_meters',
    'parks_around3000' : 'parks_around_3000_meters',
    'ponds_around3000' : 'ponds_around_3000_meters',
    'open_plan' : 'is_open_plan',
    'rooms' : 'total_rooms',
})

df.columns # проверка названий столбцов
```

```python
# приводм столбец со временем в должный вид
df['first_day_exposition'] = pd.to_datetime(
    df['first_day_exposition'], format='%Y%m%dT%H:%M:%S'
)
df.head()
```

```python
df['locality_name'] = df['locality_name'].str.lower() # приводим значения к нижнему регистру
df['locality_name'] = df['locality_name'].str.replace('ё','е') # меняем букву 'ё' на 'е'
display(df['locality_name'].sort_values().unique())
```

### Работа с пропусками

#### Работа с пропусками в булевых значениях

```python
# смотрим уникальные значения в булевых столбцах
display(df['is_apartment'].unique())
display(df['is_studio'].unique())
display(df['is_open_plan'].unique())
```

Так как частью продавцов значения столбца is_apartment не заполнялись, считаю логичным предположить, что эти объекты аппартаментами не являются

```python
# заменяем NaN на False
df['is_apartment'] = df['is_apartment'].fillna(False)
display(df['is_apartment'].unique())
```

```python
# переводим в формат int (булевые столбцы заполнятся 0 и 1 в соответствии с False и True)
df = df.astype({'is_apartment': np.int8,
                'is_studio': np.int8,
                'is_open_plan': np.int8,
})

# проверяем уникальные значения в булевых столбцах
display(df['is_apartment'].unique())
display(df['is_studio'].unique())
display(df['is_open_plan'].unique())
```

Пояснение: мой выбран тип int8, т.к. для хранения 0 и 1 не нужно выделять большой объем памяти

#### Работа с числовыми пропусками

```python
df.isna().sum() # поиск и подсчет пропусков в таблице
```

В столбцах floors_total и locality_name находится очень малое количество пропусков, удаление которых не повлияет на итоговый результат исследования

```python
df = df.dropna(subset=['floors_total', 'locality_name']).reset_index(drop=True) # удаляем малое количество пропусков
```

Пояснение про парки и водоемы: т.к. данные пункты не заполнялись продавцом, то логично предположить, что в данном радиусе они отсутствуют, по этому заполняем нулевым значением. (уточнить информацию не представляется возможным)

```python
# если пропущено, то их нет
df['balcony'] = df['balcony'].fillna(0) 
df['parks_around_3000_meters'] = df['parks_around_3000_meters'].fillna(0) 
df['ponds_around_3000_meters'] = df['ponds_around_3000_meters'].fillna(0)

# высоту потолков выставить по медианному значению
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())

# квартира-студия состоит из 1 одной комнаты, заменим все значения 0 на 1
df.loc[df['is_studio'] == 1, 'total_rooms'] = 1

# в квартире-студии кухня отсутствует, как отдельное помещение, просвоим пропускам значение 0
df.loc[df['is_studio'] == 1, 'kitchen_area'] = 0
```

Далее займемся заполнением пропуском в данных, которые необходимо определять опираясь на группы схожие по местоположению, общей площади и количестве комнат.

```python
# площадь кухни отпределить медианой, относительно площади квартиры
for i in df.loc[df['kitchen_area'].isna()]['total_area'].unique():
    df.loc[df['kitchen_area'].isna(),'kitchen_area']= (
        df.groupby('total_area')['kitchen_area'].median().reset_index().query('total_area==@i')['kitchen_area'])

# посчитать жилую площадь по медиане, относительно количества комнат
for i in df.loc[df['living_area'].isna()]['total_rooms'].unique():
    df.loc[df['living_area'].isna(),'living_area']= (
        df.groupby('total_rooms')['living_area'].median().reset_index().query('total_rooms==@i')['living_area'])

# аналогичным образом заполняем пропуски дистанций до аэропорта
for i in df.loc[df['airports_distance_meters'].isna()]['locality_name'].unique():
    df.loc[df['airports_distance_meters'].isna(),'airports_distance_meters']= (
        df.groupby('locality_name')['airports_distance_meters'].median().reset_index()
        .query('locality_name==@i')['airports_distance_meters'])
    
# аналогичным образом заполняем пропуски дистанций до аэропорта
for i in df.loc[df['city_center_distance_meters'].isna()]['locality_name'].unique():
    df.loc[df['city_center_distance_meters'].isna(),'city_center_distance_meters']= (
        df.groupby('locality_name')['city_center_distance_meters'].median().reset_index()
        .query('locality_name==@i')['city_center_distance_meters'])
    
# аналогичным образом заполняем пропуски дистанций до парков и водоемов
for i in df.loc[df['parks_distance_meters'].isna()]['locality_name'].unique():
    df.loc[df['parks_distance_meters'].isna(),'parks_distance_meters']= (
        df.groupby('locality_name')['parks_distance_meters'].median().reset_index()
        .query('locality_name==@i')['parks_distance_meters'])
    
for i in df.loc[df['ponds_distance_meters'].isna()]['locality_name'].unique():
    df.loc[df['ponds_distance_meters'].isna(),'ponds_distance_meters']= (
        df.groupby('locality_name')['ponds_distance_meters'].median().reset_index()
        .query('locality_name==@i')['ponds_distance_meters'])
    
df.isna().sum()
```

После проведенной работы по заполнению пропусков в столбцах living_area и kitchen_area все еще остались незаполненые ячейки, исходя из написанных выше функций напрашивается вывод, что в этих строках отсутствует большое количество важной информации об объектах недвижимости.
<br>Исходя из этого данные обявления можно отнести к мошейническим. Данную информацию необходимо направить в отдел, ответственный за обработку поступающей информации для улучшения алгоритмов фильтрации объявлений.
<br>Считаю логичным удалить данные строки что бы не искажать наше исследование.

Так же мы видим несколько столбцов с большим количеством пропусков:

 - airports_distance_meters;
 - city_center_distance_meters;
 - parks_distance_meters;
 - ponds_distance_meters.

<br>В иописании говорилось, что эти данные вписывают автоматически, советую оптимизировать процесс заполнения информации или процесс выгрузки данных.
<br>Думаю логично будет категоризировать эту информацию для более удобной работы с ними. Категории поместим в дополнительные столбцы нашей таблицы.

```python
def airport_distance_group(airports_distance_meters):
    if airports_distance_meters < 5001:
        return 'до 5000 м.'
    elif airports_distance_meters > 5000 and airports_distance_meters < 10001:
        return 'от 5000 м. до 10000 м.'
    elif airports_distance_meters > 10000 and airports_distance_meters < 25001:
        return 'от 10000 м. до 25000 м.'
    elif airports_distance_meters > 25000 and airports_distance_meters < 50001:
        return 'от 25000 м. до 50000 м.'
    elif airports_distance_meters > 50000:
        return 'от 50000 м.'
    else:
        return 'неизвестно'

df['grouped_airport_distance'] = df['airports_distance_meters'].apply(airport_distance_group)


def city_center_distance_group(city_center_distance_meters):
    if city_center_distance_meters < 501:
        return 'до 500 м.'
    elif city_center_distance_meters > 500 and city_center_distance_meters < 1001:
        return 'от 500 м. до 1000 м.'
    elif city_center_distance_meters > 1000 and city_center_distance_meters < 5001:
        return 'от 1000 м. до 5000 м.'
    elif city_center_distance_meters > 5000 and city_center_distance_meters < 10001:
        return 'от 5000 м. до 10000 м.'
    elif city_center_distance_meters > 10000 and city_center_distance_meters < 25001:
        return 'от 10000 м. до 25000 м.'
    elif city_center_distance_meters > 25000 and city_center_distance_meters < 50001:
        return 'от 25000 м. до 50000 м.'
    elif city_center_distance_meters > 50000:
        return 'от 50000 м.'
    else:
        return 'неизвестно'

df['grouped_city_center_distance'] = df['city_center_distance_meters'].apply(city_center_distance_group)


def parks_distance_group(parks_distance_meters):
    if parks_distance_meters < 251:
        return 'до 250 м.'
    elif parks_distance_meters > 500 and parks_distance_meters < 1001:
        return 'от 500 м. до 1000 м.'
    elif parks_distance_meters > 1000 and parks_distance_meters < 2001:
        return 'от 1000 м. до 2000 м.'
    elif parks_distance_meters > 2000 and parks_distance_meters < 3001:
        return 'от 2000 м. до 3000 м.'
    elif parks_distance_meters > 3000:
        return 'от 3000 м.'
    else:
        return 'неизвестно'

df['grouped_parks_distance'] = df['parks_distance_meters'].apply(parks_distance_group)


def ponds_distance_group(ponds_distance_meters):
    if ponds_distance_meters < 151:
        return 'до 150 м.'
    elif ponds_distance_meters > 300 and ponds_distance_meters < 501:
        return 'от 300 м. до 500 м.'
    elif ponds_distance_meters > 500 and ponds_distance_meters < 1001:
        return 'от 500 м. до 1000 м.'
    elif ponds_distance_meters > 1000:
        return 'от 1000 м.'
    else:
        return 'неизвестно'

df['grouped_ponds_distance'] = df['ponds_distance_meters'].apply(ponds_distance_group)

df.head()
```

```python
# переводим в формат int
df = df.astype({'floors_total': np.int64,
                'balcony': np.int64,
                'last_price': np.int64,
                'parks_around_3000_meters': np.int64,
                'ponds_around_3000_meters': np.int64,
                #'days_exposition': np.int64,
})
df.info()
```

```python
def floors_categorization(floors_total, floor):
    try:
        if floor == 1:
            return 'Первый'
        elif floor == floors_total:
            return 'Последний'
        elif floor > 1 and floor < floors_total:
            return 'Другой'
    except:
        return 'Ошибка'

df['floors_category'] = np.vectorize(floors_categorization)(df['floors_total'], df['floor'])

df.head()
```

##### Oпределение центральной зоны

```python
# строим таблицу с предложениями из Питера где нет пропусков значений расстояния до центра
spb = df.query('locality_name == "санкт-петербург"' and 'grouped_city_center_distance != "неизвестно"')

# Создаем дополнительный столбец, в котором расстояние до центра будет в километрах
# spb['city_center_distance_km']=(spb['city_center_distance_meters']/1000).round(0)

spb = spb.reset_index()
spb.loc[:, 'city_center_distance_km'] = (spb['city_center_distance_meters']/1000).round(0)

# Создаем таблицу сгруппированную на основе зависимости расстояния до центра в км. и средней стоимости объекта
spb_group = spb.groupby(['city_center_distance_km']).agg({'last_price' : 'mean'})

# Строим график
spb_group.plot(figsize = (20,6), style = 'go', markersize = 12, grid=True)
plt.title('Площади квартир', size = 14)
plt.xlabel('Расстояние от центра, км.')
plt.ylabel('Количество квартир, шт.')
plt.show()
```

Резкий спад стоимости объекта наблюдается на отметке ~9 км., этот радиус будем считать чентральным районом.
