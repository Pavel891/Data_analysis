## Задание

**Заказчик**
<br>Интернет-магазин «Стримчик», который продаёт по всему миру
компьютерные игры. Из открытых источников доступны исторические данные о
продажах игр, оценки пользователей и экспертов, жанры и платформы (например,
Xboxили PlayStation). 

**Задача**
<br>Необходимо выявить определяющие успешность игры
закономерности. Это позволит сделать ставку на потенциально популярный продукт и
спланировать рекламные кампании.

**Гипотезы для проверки**
 - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
 - Средние пользовательские рейтинги жанров Action (англ. «действие»,
экшен-игры) и Sports (англ. «спортивные соревнования») разные.

**Описание данных**
<br>Перед вами данные до 2016 года. 
<br>В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board)
— это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB
оценивает игровой контент и присваивает ему подходящую возрастную категорию.

**Описание таблицы**
 - Name — название игры
 - Platform — платформа
 - Year_of_Release — год выпуска
 - Genre — жанр игры
 - NA_sales — продажи в Северной Америке (миллионы проданных копий)
 - EU_sales — продажи в Европе (миллионы проданных копий)
 - JP_sales — продажи в Японии (миллионы проданных копий)
 - Other_sales — продажи в других странах (миллионы проданных копий)
 - Critic_Score — оценка критиков (максимум 100)
 - User_Score — оценка пользователей (максимум 10)
 - Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating
Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает
им подходящую возрастную категорию.

**Описание рейтингов ESRB**
 - «EC» («Early childhood») — «Для детей младшего возраста»: Игра подходит для детей от 3 лет и не содержит материалов, которые родители могли бы счесть неподходящими. Продукты, получившие данный рейтинг, изначально разрабатываются для детей и обычно представляют собой развивающие игры. Некоторые усложнённые развивающие игры могут иметь рейтинг «Everyone».
 - «E» («Everyone») — «Для всех»: Содержание вполне подходит для возрастной категории от 6 лет. Такие игры могут понравиться и взрослым. Игры с этим рейтингом могут содержать минимальное насилие, в основном «мультяшного» характера. Первой игрой, которая получила данный рейтинг, стала The Simpsons Cartoon Studio, выпущенная в 1996 году. Первоначально "K-A" ("Kids to Adults")
 - «E10+» («Everyone 10 and older») — «Для всех от 10 лет и старше»: Проекты с данным рейтингом могут содержать или немного больше мультипликационного или мягкого насилия, или несколько откровенные сцены, или минимальное количество крови. Рейтинг был принят ESRB 2 марта 2005 года. Первой игрой, которая получила данный рейтинг, стала Donkey Kong: Jungle Beat.
 - «T» («Teen») — «Подросткам»: Игра подходит для лиц от 13 лет. Проекты из данной категории могут содержать насилие, непристойные сцены, грубый юмор, в меру откровенное сексуальное содержимое, кровь или нечастое использование ненормативной лексики.
 - «M» («Mature») — «Для взрослых»: Материалы игры не подходят для подростков младше 17 лет. Проекты с данным рейтингом могут содержать достаточно жестокое насилие, большое количество крови с расчленением, непристойные сексуальные сцены или грубую ненормативную лексику, нежелательную для младшей аудитории.
 - «AO» («Adults Only 18+») — «Только для взрослых»: Содержание игры только для взрослых старше 18 лет. Продукты из данной категории могут содержать длительные сцены жестокого насилия и/или очень откровенное сексуальное содержимое, а также сцены с обнажением. Большинство таких игр предназначено для персональных компьютеров под управлением Microsoft Windows и Apple Macintosh. Рейтинг «Только для взрослых» является предметом многочисленных дискуссий, так как накладывает серьёзные ограничения на продажи игры. Первой игрой, которая получила данный рейтинг, стала The Joy of Sex, выпущенная в 1993 году.
 - «RP» («Rating Pending») — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга. Данный логотип используется только на рекламных презентациях и в демо-версиях игр до официальной даты выпуска в продажу.

**Внимание!**
<br>Данные за 2016 год могут быть неполными.

## <a id="preprocessing">Предобработка данных</a>

```python
df.duplicated().sum()
```

```python
df.columns = df.columns.str.lower()

for cols in df[['name', 'platform', 'genre', 'rating']]:
	df[cols] = df[cols].str.lower()
```

```python
df.isna().sum()
```

```python
# всего 1.6% в столбце 'year_of_release' пропусков, можно ими пожертвовать
df = df.dropna(subset=['year_of_release'])

# так же удалим пропуски в столбцах name и genre
df = df.dropna(subset=['name', 'genre'])

df.isna().sum()
```

```python
# изменяем тип данных на int
df = df.astype({'year_of_release': np.int64})

df.info()
```

Аббревиатура 'tbd'(To be determined) означает, что значение еще не назначено. Из этого определения следует, что это значение можно заменить значением NaN.

```python
# user_score нужно привести к типу float, но для этого значение tbd нужно заменить на NaN
df['user_score'] = df['user_score'].replace('tbd', np.nan)
df = df.astype({'user_score': np.float64, 'critic_score': np.float64}).round(2)

df.info()
```

```python
# Запишем в новый столбец суммарные продажи по всем регионам.
df['total_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']

df.head(15)
```

#### Вывод

В ходе предобработки мы привели данные к нужным типам данных.
<br>Удалили значения NaN в столбцах 'year_of_release', 'name', 'genre'.
<br>Оставшиеся пропуски заполнять не имеет смысла, т.к. оценки вещь весьма субъективная, особенно оценки игроков, а рейтинг могли или не назначить или пришлось бы разбирать каждую игру по отдельности, а это долго.

Причинами пропусков, на мой взгляд, являются:
 - в колонках genre и name, скорее всего, пропуски появились по техническим причинам, т.к. даже если игра только на этапе анонса, ее название и жанр называют сразу;
 - в колонке year_of_realese кроме технической причины, может быть так, что игру анонсировали, но точного времени выхода не огласили(или как в случае студии CD Project Red дату указывать не имеет смысла, даже если ее назвали, из-за привычки студии несколько раз переносить релиз);
 - в колонках, которые содержат оценки рейтинг вариантов два: техническая проблема, отсутсвие оценок и рейтингов.

### <a id="3.3">Анализ влияния оценок на продажи</a>

Давайте рассмотрим уровень влияния оценок критиков и игроков на продаваемость игр.

```python
# напишем небольшую фукцию.

def sales_and_score(game_cons, score):
	game_cons.plot(x=score, y='total_sales', kind='scatter', grid=True, alpha=0.3)
	plt.title('Влияние оценки на продажи')
	plt.xlabel('Оценка')
	plt.ylabel('Мировые продажи, млн.коп.')
	plt.show()
	print('Корреляция =', game_cons[score].corr(df['total_sales']).round(2))
```

#### 3.4.1 Оценка критиков на платформе PS3

```python
sales_and_score(df_top5_platf_filtered.query('platform == "ps3"'), 'critic_score')
```

#### 3.4.2 Оценка игроков на платформе PS3

```python
sales_and_score(df_top5_platf_filtered.query('platform == "ps3"'), 'user_score')
```

#### 3.4.3 Оценка критиков на других платформах

```python
sales_and_score(df_top5_platf_filtered.query('platform != "ps3"'), 'critic_score')
```

#### 3.4.3 Оценка игроков на других платформах

```python
sales_and_score(df_top5_platf_filtered.query('platform != "ps3"'), 'user_score')
```

Из исследования можно сделать вывод, что оценки критиков читают только еденицы игроков, владеющих платформой PS3(Наверное японцы). В остальном можно сказать, что оценки игроков и критиков практически не влияют на продажи игр.

## <a id="verification">Проверка гипотез</a>

### <a id="5.1">Гипотеза о различии пользовательских рейтингов на платформах Xbox ONE и PC</a>

H0: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые
<br>H1: Средние пользовательские рейтинги платформ Xbox One и PC разные

```python
metric_table_platf = df[df['platform'].isin(['xone', 'pc'])]
metric_table_platf = metric_table_platf.dropna(subset=['user_score'])
metric_table_platf = metric_table_platf[metric_table_platf['year_of_release'] > 2012]

variance_xone = np.var(metric_table_platf[metric_table_platf['platform'] == 'xone']['user_score'], ddof=1)
display(variance_xone)
variance_pc = np.var(metric_table_platf[metric_table_platf['platform'] == 'pc']['user_score'], ddof=1)
display(variance_pc)
```

```python

pval_platf = st.ttest_ind(metric_table_platf.query('platform == "xone"')['user_score']\
	, metric_table_platf.query('platform == "pc"')['user_score'], equal_var=False)
print('p-значение:', pval_platf.pvalue)

alpha = 0.05
if pval_platf.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
```

На мой взгляд такой результат обусловлен тем, что портирование приложений между данными платформаи не представляет особой сложности.
<br>Пояснить это можно тем, что PC для гейминга используют ОС Windows, которая так же как и платформа Xbox ONE является продуктом компании Microsoft.

### <a id="5.2">Гипотеза о различии пользовательских рейтингов жанров Action и Sports</a>

H0: Средние пользовательские рейтинги жанров Action и Sports одинаковые
<br>H1: Средние пользовательские рейтинги жанров Action и Sports разные

```python
metric_table_genres = df[df['genre'].isin(['action', 'sports'])]
metric_table_genres = metric_table_genres.dropna(subset=['user_score'])
metric_table_genres = metric_table_genres[metric_table_genres['year_of_release'] > 2012]

variance_action = np.var(metric_table_genres[metric_table_genres['genre'] == 'action']['user_score'], ddof=1)
display(variance_action)
variance_sports = np.var(metric_table_genres[metric_table_genres['genre'] == 'sports']['user_score'], ddof=1)
display(variance_sports)
```

```python
pval_genres = st.ttest_ind(metric_table_genres.query('genre == "action"')['user_score']\
	, metric_table_genres.query('genre == "sports"')['user_score'])
print('p-значение:', pval_genres.pvalue)

alpha = 0.05
if pval_genres.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
```

Этот результат вполне логичен.
